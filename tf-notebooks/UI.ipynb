```python
import sys

import tensorflow as tf
import numpy as np
from tensorflow.contrib import rnn
import PyQt5.QtWidgets
from PyQt5.QtGui import *
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *

myApp = QApplication(sys.argv)
myApp.setWindowIcon(QIcon('icon.png'))

win = QTabWidget()
win.setWindowTitle('Compose')
win.resize(940, 670)
win.setMaximumSize(1280, 720)
win.setMinimumSize(940, 670)
win.setStyleSheet("background-image:url(./background.png)")

back = QWidget(win)
back.setGeometry(10,10,920,200)
back.setStyleSheet("background-image:url(./back)")

#탭 배경 설정
myApp.setWindowIcon(QIcon('icon.png'))

#학습할때 사용할 변수
input_value = {}
genre =""
musicSheet =''
tempo =""
key = ""
checker = 0


#약보 선언
note = {}
note_key = {}
note_i=0
note_x = 20
note_y = 220
for note_i in range (121):
    if(note_x > 900):
        note_x = 20
        note_y += 110
        note[note_i] = QWidget(win)
        note[note_i].setGeometry(note_x, note_y, 30, 110)
        note[note_i].setStyleSheet("background-image:url(./note/nothing)")
        note_key[note_i] = QLabel(win)
        note_key[note_i].setGeometry(note_x, note_y, 30, 20)
        note_key[note_i].setStyleSheet("background-image:url(./null.png); color:#000000; font:bold; font-size:15px")
        note_x += 30
    elif((note_i == 2)):
        note[note_i] = QWidget(win)
        note[note_i].setGeometry(note_x, note_y, 30, 65)
        note[note_i].setStyleSheet("background-image:url(./note/nothing)")
        note_y += 65
    elif((note_i == 3)):
        note[note_i] = QWidget(win)
        note[note_i].setGeometry(note_x, note_y, 30, 45)
        note[note_i].setStyleSheet("background-image:url(./note/nothing)")
        note_x += 30
        note_y -= 65
    else:
        note[note_i] = QWidget(win)
        note[note_i].setGeometry(note_x, note_y, 30, 110)
        note[note_i].setStyleSheet("background-image:url(./note/nothing)")
        if(note_i > 3):
            note_key[note_i] = QLabel(win)
            note_key[note_i].setGeometry(note_x, note_y, 30, 20)
            note_key[note_i].setStyleSheet("background-image:url(./null.png); color:#000000; font:bold; font-size:15px")
        note_x += 30

#버튼 선언
btn = {}
v_btn= {}
btn_i=-1
btn_x = 20
btn_y = 20
while(btn_i<16):
    btn_i += 1
    if((btn_i%7) == 0):
        btn[btn_i] = QPushButton("\n\n\nC",win)
    elif((btn_i % 7) == 1):
        btn[btn_i] = QPushButton("\n\n\nD", win)
    elif((btn_i % 7) == 2):
        btn[btn_i] = QPushButton("\n\n\nE",win)
    elif((btn_i % 7) == 3):
        btn[btn_i] = QPushButton("\n\n\nF",win)
    elif((btn_i % 7) == 4):
        btn[btn_i] = QPushButton("\n\n\nG",win)
    elif((btn_i % 7) == 5):
        btn[btn_i] = QPushButton("\n\n\nA",win)
    elif((btn_i % 7) == 6):
        btn[btn_i] = QPushButton("\n\n\nB",win)

    btn[btn_i].setGeometry(btn_x, btn_y, 47, 180)
    btn[btn_i].setStyleSheet("background-image:url(./pulled); color:#000000; font:bold; font-size:30px")
    v_btn[btn_i] = False
    btn_x += 48

btn_x = -39
btn_y = 20

while(btn_i<28):
    btn_i += 1
    if((btn_i%5) == 2):
        btn[btn_i] = QPushButton("\n\nDb\nC#",win)
        btn_x += 84
    elif((btn_i % 5) == 3):
        btn[btn_i] = QPushButton("\n\nEb\nD#", win)
        btn_x += 57
    elif((btn_i % 5) == 4):
        btn[btn_i] = QPushButton("\n\nGb\nF#",win)
        btn_x += 86
    elif((btn_i % 5) == 0):
        btn[btn_i] = QPushButton("\n\nAb\nG#",win)
        btn_x += 53
    elif((btn_i % 5) == 1):
        btn[btn_i] = QPushButton("\n\nBb\nA#",win)
        btn_x += 53

    btn[btn_i].setGeometry(btn_x, btn_y, 36, 110)
    btn[btn_i].setStyleSheet("background-image:url(./pulled2); color:#ffffff; font:bold; font-size:17px")
    v_btn[btn_i] = False

sig_first = QTextEdit(win)
sig_first.setGeometry(850,70,50,48)
sig_first.setStyleSheet("background-image:url(./pulled2); color:#ffffff; font:bold; font-size:25pt")
sig_first.setReadOnly(True)
sig_first.setText("1")

sig_second = QTextEdit(win)
sig_second.setGeometry(850,113,50,48)
sig_second.setStyleSheet("background-image:url(./pulled2); color:#ffffff; font:bold; font-size:25pt")
sig_second.setReadOnly(True)
sig_second.setText("8")

up_first = QPushButton(win)
up_first.setGeometry(900, 72, 19, 19)
up_first.setStyleSheet("background-image:url(./up)")

down_first = QPushButton(win)
down_first.setGeometry(900, 91, 19, 19)
down_first.setStyleSheet("background-image:url(./down)")

up_second = QPushButton(win)
up_second.setGeometry(900, 115, 19, 19)
up_second.setStyleSheet("background-image:url(./up)")

down_second = QPushButton(win)
down_second.setGeometry(900, 134, 19, 19)
down_second.setStyleSheet("background-image:url(./down)")

w1 = QWidget(win)
w1.setGeometry(850,64,69,8)
w1.setStyleSheet("background-image:url(./pulled)")

w2 = QWidget(win)
w2.setGeometry(850,110,69,5)
w2.setStyleSheet("background-image:url(./pulled)")

w3 = QWidget(win)
w3.setGeometry(850,153,69,8)
w3.setStyleSheet("background-image:url(./pulled)")

#compose button
compose_btn = QPushButton("Compose", win)
compose_btn.setGeometry(270, 580, 100, 40)
compose_btn.setStyleSheet("background-image:url(./pulled); color:#000000; font:bold")

groupBox = QGroupBox("Genre", win)
groupBox.setGeometry(180, 555, 80, 105)
groupBox.setStyleSheet("background-image:url(./note/nothing); color:#ffffff; font:bold; font-size:17px")

radio1 = QRadioButton("Jigs", win)
radio1.setGeometry(190, 610, 60, 20)
radio1.setChecked(True)
radio1.setStyleSheet("background-image:url(./note/nothing); color:#ffffff; font:bold")

radio2 = QRadioButton("Waltz", win)
radio2.setGeometry(190, 630, 60, 20)
radio2.setStyleSheet("background-image:url(./note/nothing); color:#ffffff; font:bold")

compose_txt = QTextEdit(win)
compose_txt.setGeometry(190, 580, 60, 25)
compose_txt.setStyleSheet("background-image:url(./note/nothing); color:#000000; font:bold")

store_btn = QPushButton("Store", win)
store_btn.setGeometry(550, 580, 100, 40)
store_btn.setStyleSheet("background-image:url(./pulled); color:#000000; font:bold")

@pyqtSlot()
def on_up_first():
    str_f = sig_first.toPlainText()
    str_i = int(str_f)+1
    if(str_i >= 13):
        str_i = 1
    str_f = str(str_i)
    sig_first.setText(str_f)
@pyqtSlot()
def on_down_first():
    str_f = sig_first.toPlainText()
    str_i = int(str_f)-1
    if(str_i <= 0):
        str_i = 12
    str_f = str(str_i)
    sig_first.setText(str_f)

@pyqtSlot()
def on_up_second():
    str_s = sig_second.toPlainText()
    str_i = int(str_s)+1
    if(str_i >= 13):
        str_i = 1
    str_s = str(str_i)
    sig_second.setText(str_s)

@pyqtSlot()
def on_down_second():
    str_s = sig_second.toPlainText()
    str_i = int(str_s) - 1
    if (str_i <= 0):
        str_i = 12
    str_s = str(str_i)
    sig_second.setText(str_s)

@pyqtSlot(int)
def on_btn(a):
    for i in range (29):
        v_btn[i] = False
        if(i < 17):
            btn[i].setStyleSheet("background-image:url(./pulled); color:#000000; font:bold; font-size:30px")
        else:
            btn[i].setStyleSheet("background-image:url(./pulled2); color:#ffffff; font:bold; font-size:17px")
    v_btn[a]= True
    input_value[0] = str(a)

    if (a < 17):
        btn[a].setStyleSheet("background-image:url(./pushed); color:#ffffff; font:bold; font-size:30px")
    else:
        btn[a].setStyleSheet("background-image:url(./pushed); color:#ffffff; font:bold; font-size:17px")

def drawdot(musicSheet):
    d_v = 5
    d_n = 0
    
    musicSheet+='     '
    while(d_n < len(musicSheet) and d_v < 91):
        if(d_v == 31 or d_v == 61):
            note[d_v].setStyleSheet("background-image:url(./note/start)")
            d_v += 1
            d_n -= 1
        elif(musicSheet[d_n] =='"' and musicSheet[d_n+2]=='"'):
            note_key[d_v].setText(musicSheet[d_n+1])
            d_n += 2
        elif(musicSheet[d_n] =='"' and musicSheet[d_n+3]=='"'):
            note_key[d_v].setText(musicSheet[d_n+1]+musicSheet[d_n+2])
            d_n += 3
        elif(musicSheet[d_n] == 'A'):
            if(musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/407)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/807)")
            d_v += 1

        elif(musicSheet[d_n] == 'B'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/408)")
                d_n += 1
            elif(musicSheet[d_n + 1] == ','):
                if(musicSheet[d_n + 2] == '2'):
                    note[d_v].setStyleSheet("background-image:url(./note/401)")
                    d_n += 2
                else:
                    note[d_v].setStyleSheet("background-image:url(./note/801)")
                    d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/808)")
            d_v += 1
        elif (musicSheet[d_n] == 'C'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/402)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/802)")
            d_v += 1

        elif (musicSheet[d_n] == 'D'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/403)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/803)")
            d_v += 1

        elif (musicSheet[d_n] == 'E'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/404)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/804)")
            d_v += 1

        elif (musicSheet[d_n] == 'F'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/405)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/805)")
            d_v += 1

        elif (musicSheet[d_n] == 'G'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/406)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/806)")
            d_v += 1

        elif(musicSheet[d_n] == 'c'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/409)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/809)")
            d_v += 1

        elif(musicSheet[d_n] == 'd'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/410)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/810)")
            d_v += 1

        elif(musicSheet[d_n] == 'e'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/411)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/811)")
            d_v += 1

        elif(musicSheet[d_n] == 'f'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/412)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/812)")
            d_v += 1

        elif(musicSheet[d_n] == 'g'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/413)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/813)")
            d_v += 1

        elif(musicSheet[d_n] == 'a'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/414)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/814)")
            d_v += 1

        elif(musicSheet[d_n] == 'b'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/415)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/815)")
            d_v += 1

        elif (musicSheet[d_n] == 'z'):
            if (musicSheet[d_n + 1] == '2'):
                note[d_v].setStyleSheet("background-image:url(./note/z2)")
            else:
                note[d_v].setStyleSheet("background-image:url(./note/z)")
            d_v += 1

        elif (musicSheet[d_n] == '|'):
            if (musicSheet[d_n + 1] == '|'):
                note[d_v].setStyleSheet("background-image:url(./note/stick_2)")
                d_n += 1
            else:
                note[d_v].setStyleSheet("background-image:url(./note/stick_1)")
            d_v += 1

        elif (musicSheet[d_n] == ':'):
            if (musicSheet[d_n + 1] == '|'):
                note[d_v].setStyleSheet("background-image:url(./note/stick_3)")
            d_v += 1
            d_n += 1
        
        else: 
            note[d_v].setStyleSheet("background-image:url(./note/default_note)")
            d_v += 1
        
        d_n += 1

           
    if(d_v < 31 and d_v > 1):
        while(d_v<31):
            note[d_v].setStyleSheet("background-image:url(./note/default_note)")
            d_v += 1
    elif(d_v < 61 and d_v > 31):
        while(d_v<61):
            note[d_v].setStyleSheet("background-image:url(./note/default_note)")
            d_v += 1
    elif(d_v < 91 and d_v > 61):
        while(d_v<91):
            note[d_v].setStyleSheet("background-image:url(./note/default_note)")
            d_v += 1
    elif(d_v < 121 and d_v > 91):
        while(d_v<121):
            note[d_v].setStyleSheet("background-image:url(./note/default_note)")
            d_v += 1

@pyqtSlot()
def on_str_btn():
    store_path = "./temp.txt"
    wf = open(store_path, 'w')
    wf.write(musicSheet)
    wf.close()


@pyqtSlot()
def on_com_btn():
    flag = False
    global key
    global genre
    global tempo
    global musicSheet
    global checker
    
#input 구별 genre
    fgen = 0
    if radio1.isChecked():
        genre = "jigs"
    elif radio2.isChecked():
        genre = "walts"

    while(fgen<29):
        if(v_btn[fgen] == True):
            flag = True
            break;
        else:
            fgen+=1
# input 구별 key
    if(flag == True):
        str_f = sig_first.toPlainText()
        str_s = sig_second.toPlainText()
        note[0].setStyleSheet("background-image:url(./note/start)")
        if (fgen < 16):
            if ((fgen % 7) == 0):
                note[1].setStyleSheet("background-image:url(./note/default_note)")
                key = "C"
            elif ((fgen % 7) == 1):
                note[1].setStyleSheet("background-image:url(./note/2_key)")
                key = "D"
            elif ((fgen % 7) == 2):
                note[1].setStyleSheet("background-image:url(./note/4_key)")
                key = "E"
            elif ((fgen % 7) == 3):
                note[1].setStyleSheet("background-image:url(./note/1_bkey)")
                key = "F"
            elif ((fgen % 7) == 4):
                note[1].setStyleSheet("background-image:url(./note/1_key)")
                key = "G"
            elif ((fgen % 7) == 5):
                note[1].setStyleSheet("background-image:url(./note/3_key)")
                key = "A"
            elif ((fgen % 7) == 6):
                note[1].setStyleSheet("background-image:url(./note/5_key)")
                key = "B"
        elif(f < 28):
            if ((fgen % 5) == 2):
                note[1].setStyleSheet("background-image:url(./note/7_key)")
                key = "Db"
            elif ((fgen % 5) == 3):
                note[1].setStyleSheet("background-image:url(./note/3_bkey)")
                key = "Eb"
            elif ((fgen % 5) == 4):
                note[1].setStyleSheet("background-image:url(./note/6_key)")
                key = "Gb"
            elif ((fgen % 5) == 0):
                note[1].setStyleSheet("background-image:url(./note/4_bkey)")
                key = "Ab"
            elif ((fgen % 5) == 1):
                note[1].setStyleSheet("background-image:url(./note/2_bkey)")
                key = "Bb"

        note[2].setStyleSheet("background-image:url(./note/"+ str_f + "_up)")
        note[3].setStyleSheet("background-image:url(./note/"+ str_s + "_down)")
        note[4].setStyleSheet("background-image:url(./note/default_note)")

#input 구별 tempo
        tempo = str_f + "/" + str_s
        filename = genre + ".csv"

        msgBox = QMessageBox()
        msgBox.setStandardButtons(QMessageBox.Close)
        msgBox.setText("기존의 데이터가 존재하는 경우 데이터가 손실될 수 있습니다.\n 새로 학습하시겠습니까?")
        yesBtn = msgBox.addButton('Yes', QMessageBox.ActionRole)
        noBtn = msgBox.addButton('No', QMessageBox.ActionRole)
        ret = msgBox.exec()


        
        
#input 구별 newLearn
        newLearn = 0
        if ret == QMessageBox.Close:
            QCloseEvent()  # should close the app, but it closes the message box
        elif msgBox.clickedButton() == yesBtn:
            newLearn = 1
        elif msgBox.clickedButton() == noBtn:
            newLearn = 0
        #학습모델 저장, 로드
        # 학습 데이터 저장.

        def save(self, filename):
            saver = tf.train.Saver()
            saver.save(sess, filename)
        def load(self, filename):
            saver = tf.train.Saver()
            saver.restore(sess, tf.train.latest_checkpoint(filename))
        
        with open(filename, ) as f:
            a = f.read().split('I')

        song = list()

        for i in range(len(a)):
            b = a[i].split(',')
            c = list()

            for j in range(len(b)):
                c.append(b[j])

            # 박자 입력
            if (c[1] == tempo):

                # 코드상에 콤마(,)가 있을 경우
                if (len(c) > 6):
                    for k in range(len(c) - 6):
                        c[5] = c[5] + ',' + c[6]
                        del c[6]

                c[5] = str(c[5]).replace('""', '"')
                c[5] = c[5][1:-2]
                song.append(c)

        print(len(song))
        print(song)

        # In[6]:
        # 코드 문자셋
        # 코드에서 사용한 문자열를 하나하나 번호를 붙혀
        # 딕셔너리로 관리
        chord_set = []

        for i in range(len(song) - 1):
            chord_set.extend(list(set(song[i + 1][5])))

        chord_dic = {w: i for i, w in enumerate(chord_set)}

        print(chord_set)
        print(chord_dic)

        # In[7]:
        # RNN 함수에 들어갈 매개변수 생성
        tf.reset_default_graph()  
        data_dim = len(chord_set)
        hidden_size = len(chord_set)
        num_classes = len(chord_set)
        sequence_length = 10  # Any arbitrary number
        learning_rate = 0.1

        # In[8]:
        # 학습 시킬 데이터의 입력과 출력 리스트 선언

        dataX = []
        dataY = []

        # 그래프 노드
        with tf.name_scope("Wx") as scope:
            y = tf
        # In[9]:
        # 1마디(문자 10개) 기준으로 선언한 리스트에 데이터 입력
        for j in range(len(song) - 1):
            for i in range(0, len(song[j + 1][5]) - sequence_length):
                x_str = song[j + 1][5][i:i + sequence_length]
                y_str = song[j + 1][5][i + 1: i + sequence_length + 1]

                print(i, x_str, '->', y_str)

                x = [chord_dic[c] for c in x_str]
                y = [chord_dic[c] for c in y_str]

                dataX.append(x)
                dataY.append(y)

        # In[10]:
        # 입력시킬 한번에 dataX만큼 학습시킴(batch_size)
        batch_size = len(dataX)
        X = tf.placeholder(tf.int32, [None, sequence_length])
        Y = tf.placeholder(tf.int32, [None, sequence_length])

        print(batch_size)
        
        # In[11]:
        # one hot 생성
        X_one_hot = 0
        X_one_hot = tf.one_hot(X, num_classes)
        print(X_one_hot)

        # In[12]:


        # lstm 함수

        def lstm_cell():
            cell = rnn.BasicLSTMCell(hidden_size, state_is_tuple=True)
            return cell

        # In[13]:
        # 셀 선언
        multi_cells = rnn.MultiRNNCell([lstm_cell() for _ in range(2)], state_is_tuple=True)

        # In[14]:
    
        #if checker == 0 : 
        #    checker += 1
        outputs, _states = tf.nn.dynamic_rnn(multi_cells, X_one_hot, dtype=tf.float32)
            
        # In[15]:
        X_for_fc = tf.reshape(outputs, [-1, hidden_size])
        outputs = tf.contrib.layers.fully_connected(X_for_fc, num_classes, activation_fn=None)

        # In[16]:
        outputs = tf.reshape(outputs, [batch_size, sequence_length, num_classes])

        # In[17]:
        weights = tf.ones([batch_size, sequence_length])
        # In[18]:


        sequence_loss = tf.contrib.seq2seq.sequence_loss(logits=outputs, targets=Y, weights=weights)

        # In[19]:


        mean_loss = tf.reduce_mean(sequence_loss)
        train_op = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(mean_loss)

        # In[20]:


        sess = tf.Session()
        sess.run(tf.global_variables_initializer())

        # In[29]:


        # 학습
        loadDataroot = "./" + genre + '/'
        saveDataroot = loadDataroot + genre + ".pd"
        # 새로 학습을 하는 경우
        str_f = sig_first.toPlainText()

        if int(newLearn) != 0:
            eEdit = compose_txt.toPlainText()
            eduNum = int(eEdit)
            for i in range(eduNum):
                _, l, results = sess.run(
                    [train_op, mean_loss, outputs], feed_dict={X: dataX, Y: dataY})
                for j, result in enumerate(results):
                    index = np.argmax(result, axis=1)
                    if i % 50 == 0:
                        print(i, j, ''.join([chord_set[t] for t in index]), l)

            # 학습내용 저장
            save(sess, saveDataroot)

        # 기존 학습 내용을 불러오는 경우
        else:
            # 학습파일 load
            try:

                load(sess, loadDataroot)
                print("학습완료")
            # 없을경우 새로 학습
            except:
                eduNum = input("학습된 장르가 아닙니다. 학습을 시작합니다. 횟수 설정해주세요.")
                for i in range(int(eduNum)):
                    _, l, results = sess.run(
                        [train_op, mean_loss, outputs], feed_dict={X: dataX, Y: dataY})
                    for j, result in enumerate(results):
                        index = np.argmax(result, axis=1)
                        if i % 50 == 0:
                            print(i, j, ''.join([chord_set[t] for t in index]), l)

                # 학습내용 저장
                save(sess, saveDataroot)

        # In[30]:
        # 결과 출력
        strtempo = "M:" + tempo 
        strL = "L:1/8"
        strK = "K:" + key 

        results = sess.run(outputs, feed_dict={X: dataX})

        musiclist = list()
   
        print(strtempo)
        print(strL)
        print(strK)

        for j, result in enumerate(results):
            index = np.argmax(result, axis=1)
            if j is 0:  # print all for the first result to make a sentence
                #print(''.join([chord_set[t] for t in index]), end='')
                musiclist.append(''.join([chord_set[t] for t in index]))
            else:
                #print(chord_set[index[-1]], end='')
                musiclist.append(chord_set[index[-1]])

        # In[27]:


        # 출력 결과 string으로 저장 .
        musicSheet = "".join(str(x) for x in musiclist)

        # In[28]:

        print(musicSheet)
        drawdot(musicSheet)
        musiclist.insert(0,strK)
        musiclist.insert(0,strL)
        musiclist.insert(0,strtempo)
        
        musicSheet = "".join(str(x) for x in musiclist)


    else:
        com_msg = QMessageBox()
        com_msg.setGeometry(700,400, 100, 100)
        com_msg.information(com_msg, "Warning", "\nKey를 선택해주세요.\n")


    

#코드 버튼
btn[0].clicked.connect(lambda: on_btn(0))
btn[1].clicked.connect(lambda: on_btn(1))
btn[2].clicked.connect(lambda: on_btn(2))
btn[3].clicked.connect(lambda: on_btn(3))
btn[4].clicked.connect(lambda: on_btn(4))
btn[5].clicked.connect(lambda: on_btn(5))
btn[6].clicked.connect(lambda: on_btn(6))
btn[7].clicked.connect(lambda: on_btn(7))
btn[8].clicked.connect(lambda: on_btn(8))
btn[9].clicked.connect(lambda: on_btn(9))
btn[10].clicked.connect(lambda: on_btn(10))
btn[11].clicked.connect(lambda: on_btn(11))
btn[12].clicked.connect(lambda: on_btn(12))
btn[13].clicked.connect(lambda: on_btn(13))
btn[14].clicked.connect(lambda: on_btn(14))
btn[15].clicked.connect(lambda: on_btn(15))
btn[16].clicked.connect(lambda: on_btn(16))
btn[17].clicked.connect(lambda: on_btn(17))
btn[18].clicked.connect(lambda: on_btn(18))
btn[19].clicked.connect(lambda: on_btn(19))
btn[20].clicked.connect(lambda: on_btn(20))
btn[21].clicked.connect(lambda: on_btn(21))
btn[22].clicked.connect(lambda: on_btn(22))
btn[23].clicked.connect(lambda: on_btn(23))
btn[24].clicked.connect(lambda: on_btn(24))
btn[25].clicked.connect(lambda: on_btn(25))
btn[26].clicked.connect(lambda: on_btn(26))
btn[27].clicked.connect(lambda: on_btn(27))
btn[28].clicked.connect(lambda: on_btn(28))


#박자 버튼
up_first.clicked.connect(on_up_first)
down_first.clicked.connect(on_down_first)
up_second.clicked.connect(on_up_second)
down_second.clicked.connect(on_down_second)

#실행 버튼
compose_btn.clicked.connect(on_com_btn)

#저장 버튼
store_btn.clicked.connect(on_str_btn)

win.show()
myApp.exec_()
```
